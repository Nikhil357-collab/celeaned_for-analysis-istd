```python
# Remove Critical Missing Values
df = df.dropna(subset=["InvoiceNo", "InvoiceDate", "CustomerID", "Quantity", "UnitPrice"])
```

```python
#Remove Duplicate Transactions Based on Key Fields
df = df.drop_duplicates(subset=["InvoiceNo", "StockCode", "CustomerID"])
```

```python
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"], errors="coerce")
df["CustomerID"] = df["CustomerID"].astype(str)
```

```python
#Remove Cancelled Orders Invoices starting with "C" indicate cancellations.
df = df[~df["InvoiceNo"].str.startswith("C")]
```

```python
#Standardize Text Columns
df["Description"] = df["Description"].str.strip().str.lower()
df["Country"] = df["Country"].str.strip().str.title()
```

```python
#Identify Outliers

Q1 = df["Revenue"].quantile(0.25)
Q3 = df["Revenue"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
#extreame outlier remove
df = df[(df["Revenue"] >= lower_bound) & (df["Revenue"] <= upper_bound)]
```

```python
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["Day"] = df["InvoiceDate"].dt.day
df["DayOfWeek"] = df["InvoiceDate"].dt.day_name()
```

```python
#Create Order-Level Revenue (Optional
order_revenue = df.groupby("InvoiceNo")["Revenue"].sum().reset_index()
```

```python
df.to_csv("Cleaned_ecommerce_data.csv", index=False)
```
